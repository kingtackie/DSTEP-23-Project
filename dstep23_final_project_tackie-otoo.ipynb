# **A SPATIAL-TEMPORAL ANALYSIS OF COMPLAINT AND ARREST DATA AND EVALUATING THE FAIRNESS OF THE CRIMINAL JUSTICE SYSTEM IN NEW YORK CITY**

## **PROBLEM STATEMENT**

The criminal justice system in New York City is a complex and dynamic system which plays a pivotal part in maintaining public safety and enforcing societal norms. However, the criminal justice system has come under a lot of scrutiny in the past years over racial disparity in thier effort to maintain law and order in New York City. The height of this outcry was during the COVID-19 pandemic, when issues of racial equity in policing and the court system came under question leading to concerns regarding its effectiveness and fairness. This project aims to address these major concerns by conducting a spatial-temporal analysis of complaints and arrests in New York City across its 77 police precincts. The main goal is to identify complaint and arrest hotspots and analyse the temporal patterns that have existed in the last 10 years. The evaluation of the fairness of the criminal justice system will be done using the arrest and incarceration data, to uncover any underlying racial disparities. This will provide insights that can inform evidence-based policy recommendations for improving the effectiveness and fairness of the criminal justice system.

## **DATA**

All datasets were available at [**NYC OPEN DATA**](https://opendata.cityofnewyork.us/). The datasets for complaints and arrests are from 2006 to the end of 2022. For the purpose of this analysis, a 10-year period between 2012 to 2022 was considered. Although the incarceration starts from 2016 to date, due to data quality issues, a 5 year period from 2018 to 2022 will be used.

It is worth noting that COMPLAINTS DATA **does not** necessarily mean CRIME DATA. Complaints may not give a true reflect of actual crime. With assessibility to crime data being near impossible, complaints data **may be** the closest to giving a indication of crime, although not necessary entire true in most cases. There is a possibility of there being low complaints in area of high crime.

### **Arrest Data**

```python
# -- read in arrest data
arrest0 = pd.read_csv('https://data.cityofnewyork.us/api/views/8h9b-rp9u/rows.csv?accessType=DOWNLOAD',low_memory=False, parse_dates=['ARREST_DATE'])

# -- drop irrelevant columns
cols = ['ARREST_KEY', 'ARREST_DATE', 'ARREST_BORO', 'ARREST_PRECINCT',
       'AGE_GROUP', 'PERP_SEX', 'PERP_RACE', 'Latitude', 'Longitude']
arrest = arrest0[cols].copy()

# -- rename columns for easy use
columns = ['id','date','boro','precinct','age_group','sex','race','latitude','longitude']
arrest.columns = columns

# -- change wrong datatypes
arrest['id'] = arrest['id'].astype(str)
arrest['precinct'] = arrest['precinct'].astype(int).astype(str)

# -- drop rows with NaNs
arrest = arrest.dropna()

# -- rename race and consolidate hispanic race
arrest.loc[arrest['race']=='BLACK', 'race'] = 'black'
arrest.loc[(arrest['race']=='WHITE HISPANIC') |
           (arrest['race']=='BLACK HISPANIC'), 'race'] ='hispanic'
arrest.loc[arrest['race']=='WHITE', 'race'] = 'white'
arrest.loc[arrest['race']=='ASIAN / PACIFIC ISLANDER', 'race'] = 'asian'
arrest.loc[arrest['race']=='UNKNOWN', 'race'] = 'unknown'
arrest.loc[arrest['race']=='AMERICAN INDIAN/ALASKAN NATIVE', 'race'] = 'native'
arrest.loc[arrest['race']=='OTHER', 'race'] = 'other'

# -- change borough names from abbrev. to full name
arrest.loc[arrest['boro']=='K', 'boro'] = 'Brooklyn'
arrest.loc[arrest['boro']=='M', 'boro'] = 'Manhattan'
arrest.loc[arrest['boro']=='B', 'boro'] = 'Bronx'
arrest.loc[arrest['boro']=='Q', 'boro'] = 'Queens'
arrest.loc[arrest['boro']=='S', 'boro'] = 'Staten Island'

# -- change sex/gender from abbrev. to full name
arrest.loc[arrest['sex']=='M', 'sex'] = 'male'
arrest.loc[arrest['sex']=='F', 'sex'] = 'female'

# -- subselect data between 2012-2022
arrest = arrest[arrest['date']>'2011-12-31']

arrest
```

### **Complaints Data**


```python
# -- save data path
dpath = '/content/drive/MyDrive/Data Science Tools For Evidence Based Policy/DSTEP23 Final Project/'

# -- read in complaints data
complaint0 = pd.read_csv(dpath + 'NYPD_Complaint_Data_Historic.csv',
                        parse_dates=['RPT_DT'], low_memory=False)

# -- drop irrelevant columns
cols = ['CMPLNT_NUM', 'RPT_DT', 'BORO_NM', 'Police Precincts',
       'SUSP_AGE_GROUP', 'SUSP_RACE', 'SUSP_SEX', 'VIC_AGE_GROUP',
       'VIC_RACE', 'VIC_SEX', 'Latitude', 'Longitude']
complaint = complaint0[cols].copy()

# -- rename columns for easy use
columns = ['id','date','boro', 'precinct', 'susp_age_group',
           'susp_race', 'susp_sex', 'vic_age_group', 'vic_race',
           'vic_sex', 'latitude','longitude']
complaint.columns = columns

# -- capitalize only first letters of borough names
complaint['boro'] = complaint['boro'].str.title()

# -- drop rows with NaNs
complaint = complaint.dropna()

# # -- rename race and consolidate hispanic race
complaint.loc[complaint['susp_race']=='BLACK', 'susp_race'] = 'black'
complaint.loc[(complaint['susp_race']=='WHITE HISPANIC') |
           (complaint['susp_race']=='BLACK HISPANIC'), 'susp_race'] ='hispanic'
complaint.loc[complaint['susp_race']=='WHITE', 'susp_race'] = 'white'
complaint.loc[complaint['susp_race']=='ASIAN / PACIFIC ISLANDER', 'susp_race'] = 'asian'
complaint.loc[complaint['susp_race']=='UNKNOWN', 'susp_race'] = 'unknown'
complaint.loc[complaint['susp_race']=='AMERICAN INDIAN/ALASKAN NATIVE', 'susp_race'] = 'native'


# -- change sex/gender from abbrev. to full name
complaint.loc[complaint['susp_sex']=='M', 'susp_sex'] = 'male'
complaint.loc[complaint['susp_sex']=='F', 'susp_sex'] = 'female'

# -- subselect data between 2012-2022
complaint = complaint[complaint['date']>'2011-12-31']

# -- get correct precinct codes (using lat and lon) from the precinct shapefiles
# -- first convert complaint data to geodataframe for spatial join
geo = [Point(lon, lat) for lon, lat in zip(complaint['longitude'], complaint['latitude'])]
complaint_geo = gp.GeoDataFrame(complaint, geometry=geo, crs=precinct.crs)

# -- spatial join comapliant data to precinct data to get right precinct number
complaint = gp.sjoin(complaint_geo, precinct, how='left', predicate='within')

# -- subselect relevant columns
cols = ['id', 'date', 'boro', 'precinct_right', 'susp_age_group', 'susp_race',
       'susp_sex', 'vic_age_group', 'vic_race', 'vic_sex', 'latitude',
       'longitude']
complaint = complaint[cols]

# -- rename columns names
columns = ['id', 'date', 'boro', 'precinct', 'susp_age_group', 'susp_race',
       'susp_sex', 'vic_age_group', 'vic_race', 'vic_sex', 'latitude',
       'longitude']
complaint.columns = columns

complaint
```
