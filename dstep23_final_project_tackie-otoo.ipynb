# **A SPATIAL-TEMPORAL ANALYSIS OF COMPLAINT AND ARREST DATA AND EVALUATING THE FAIRNESS OF THE CRIMINAL JUSTICE SYSTEM IN NEW YORK CITY**

## **PROBLEM STATEMENT**

The criminal justice system in New York City is a complex and dynamic system which plays a pivotal part in maintaining public safety and enforcing societal norms. However, the criminal justice system has come under a lot of scrutiny in the past years over racial disparity in thier effort to maintain law and order in New York City. The height of this outcry was during the COVID-19 pandemic, when issues of racial equity in policing and the court system came under question leading to concerns regarding its effectiveness and fairness. This project aims to address these major concerns by conducting a spatial-temporal analysis of complaints and arrests in New York City across its 77 police precincts. The main goal is to identify complaint and arrest hotspots and analyse the temporal patterns that have existed in the last 10 years. The evaluation of the fairness of the criminal justice system will be done using the arrest and incarceration data, to uncover any underlying racial disparities. This will provide insights that can inform evidence-based policy recommendations for improving the effectiveness and fairness of the criminal justice system.

## **DATA**

All datasets were available at [**NYC OPEN DATA**](https://opendata.cityofnewyork.us/). The datasets for complaints and arrests are from 2006 to the end of 2022. For the purpose of this analysis, a 10-year period between 2012 to 2022 was considered. Although the incarceration starts from 2016 to date, due to data quality issues, a 5 year period from 2018 to 2022 will be used.

It is worth noting that COMPLAINTS DATA **does not** necessarily mean CRIME DATA. Complaints may not give a true reflect of actual crime. With assessibility to crime data being near impossible, complaints data **may be** the closest to giving a indication of crime, although not necessary entire true in most cases. There is a possibility of there being low complaints in area of high crime.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gp
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from shapely.geometry import Point
import contextily as ctx
import plotly_express as px
```

### **Geospatial Data**


```python
# -- read in borough data
boro = gp.read_file('https://data.cityofnewyork.us/api/geospatial/tqmj-j8zm?method=export&format=Shapefile')

# -- read in precint data
precinct = gp.read_file('https://data.cityofnewyork.us/api/geospatial/78dh-3ptz?method=export&format=Shapefile')
precinct['precinct'] = precinct['precinct'].astype(int).astype(str)

# -- overplot borough and precint and include basemap
fig, ax = plt.subplots(figsize=(7,7))
boro.plot('boro_name', edgecolor='k', ax=ax, legend=True,
          legend_kwds={'loc':'upper left'}, alpha=0.8)
precinct.plot(facecolor="none", edgecolor='darkred', lw=0.7, legend=True, ax=ax)
bm = ctx.add_basemap(ax, crs=boro.crs)
fig.show()
```

A map showing all the five boroughs in New York City (NYC). These boroughs are Bronx, Brooklyn, Manhattan, Queens and Staten Island. Within the boundaries of the boroughs are the precincts in NYC. Manhattan and Brooklyn have the highest number of precints. Staten Island has only four police precincts.

### **Arrest Data**

```python
# -- read in arrest data
arrest0 = pd.read_csv('https://data.cityofnewyork.us/api/views/8h9b-rp9u/rows.csv?accessType=DOWNLOAD',low_memory=False, parse_dates=['ARREST_DATE'])

# -- drop irrelevant columns
cols = ['ARREST_KEY', 'ARREST_DATE', 'ARREST_BORO', 'ARREST_PRECINCT',
       'AGE_GROUP', 'PERP_SEX', 'PERP_RACE', 'Latitude', 'Longitude']
arrest = arrest0[cols].copy()

# -- rename columns for easy use
columns = ['id','date','boro','precinct','age_group','sex','race','latitude','longitude']
arrest.columns = columns

# -- change wrong datatypes
arrest['id'] = arrest['id'].astype(str)
arrest['precinct'] = arrest['precinct'].astype(int).astype(str)

# -- drop rows with NaNs
arrest = arrest.dropna()

# -- rename race and consolidate hispanic race
arrest.loc[arrest['race']=='BLACK', 'race'] = 'black'
arrest.loc[(arrest['race']=='WHITE HISPANIC') |
           (arrest['race']=='BLACK HISPANIC'), 'race'] ='hispanic'
arrest.loc[arrest['race']=='WHITE', 'race'] = 'white'
arrest.loc[arrest['race']=='ASIAN / PACIFIC ISLANDER', 'race'] = 'asian'
arrest.loc[arrest['race']=='UNKNOWN', 'race'] = 'unknown'
arrest.loc[arrest['race']=='AMERICAN INDIAN/ALASKAN NATIVE', 'race'] = 'native'
arrest.loc[arrest['race']=='OTHER', 'race'] = 'other'

# -- change borough names from abbrev. to full name
arrest.loc[arrest['boro']=='K', 'boro'] = 'Brooklyn'
arrest.loc[arrest['boro']=='M', 'boro'] = 'Manhattan'
arrest.loc[arrest['boro']=='B', 'boro'] = 'Bronx'
arrest.loc[arrest['boro']=='Q', 'boro'] = 'Queens'
arrest.loc[arrest['boro']=='S', 'boro'] = 'Staten Island'

# -- change sex/gender from abbrev. to full name
arrest.loc[arrest['sex']=='M', 'sex'] = 'male'
arrest.loc[arrest['sex']=='F', 'sex'] = 'female'

# -- subselect data between 2012-2022
arrest = arrest[arrest['date']>'2011-12-31']

arrest
```

### **Complaints Data**


```python
# -- save data path
dpath = '/content/drive/MyDrive/Data Science Tools For Evidence Based Policy/DSTEP23 Final Project/'

# -- read in complaints data
complaint0 = pd.read_csv(dpath + 'NYPD_Complaint_Data_Historic.csv',
                        parse_dates=['RPT_DT'], low_memory=False)

# -- drop irrelevant columns
cols = ['CMPLNT_NUM', 'RPT_DT', 'BORO_NM', 'Police Precincts',
       'SUSP_AGE_GROUP', 'SUSP_RACE', 'SUSP_SEX', 'VIC_AGE_GROUP',
       'VIC_RACE', 'VIC_SEX', 'Latitude', 'Longitude']
complaint = complaint0[cols].copy()

# -- rename columns for easy use
columns = ['id','date','boro', 'precinct', 'susp_age_group',
           'susp_race', 'susp_sex', 'vic_age_group', 'vic_race',
           'vic_sex', 'latitude','longitude']
complaint.columns = columns

# -- capitalize only first letters of borough names
complaint['boro'] = complaint['boro'].str.title()

# -- drop rows with NaNs
complaint = complaint.dropna()

# # -- rename race and also consolidate hispanic race
complaint.loc[complaint['susp_race']=='BLACK', 'susp_race'] = 'black'
complaint.loc[(complaint['susp_race']=='WHITE HISPANIC') |
           (complaint['susp_race']=='BLACK HISPANIC'), 'susp_race'] ='hispanic'
complaint.loc[complaint['susp_race']=='WHITE', 'susp_race'] = 'white'
complaint.loc[complaint['susp_race']=='ASIAN / PACIFIC ISLANDER', 'susp_race'] = 'asian'
complaint.loc[complaint['susp_race']=='UNKNOWN', 'susp_race'] = 'unknown'
complaint.loc[complaint['susp_race']=='AMERICAN INDIAN/ALASKAN NATIVE', 'susp_race'] = 'native'

# -- change sex/gender from abbrev. to full name
complaint.loc[complaint['susp_sex']=='M', 'susp_sex'] = 'male'
complaint.loc[complaint['susp_sex']=='F', 'susp_sex'] = 'female'

# -- subselect data between 2012-2022
complaint = complaint[complaint['date']>'2011-12-31']

# -- get correct precinct codes (using lat and lon) from the precinct shapefiles
# -- first convert complaint data to geodataframe for spatial join
geo = [Point(lon, lat) for lon, lat in zip(complaint['longitude'], complaint['latitude'])]
complaint_geo = gp.GeoDataFrame(complaint, geometry=geo, crs=precinct.crs)

# -- spatial join comapliant data to precinct data to get right precinct number
complaint = gp.sjoin(complaint_geo, precinct, how='left', predicate='within')

# -- subselect relevant columns
cols = ['id', 'date', 'boro', 'precinct_right', 'susp_age_group', 'susp_race',
       'susp_sex', 'vic_age_group', 'vic_race', 'vic_sex', 'latitude',
       'longitude']
complaint = complaint[cols]

# -- rename columns names
columns = ['id', 'date', 'boro', 'precinct', 'susp_age_group', 'susp_race',
       'susp_sex', 'vic_age_group', 'vic_race', 'vic_sex', 'latitude',
       'longitude']
complaint.columns = columns

complaint
```

### **Incarceration Data**


```python
# -- read in incarceration data
jail = pd.read_csv('https://data.cityofnewyork.us/api/views/7479-ugqb/rows.csv?date=20231120&accessType=DOWNLOAD',
                   low_memory=False, parse_dates=['ADMITTED_DT'])

# -- drop irrelevant columns
cols = ['INMATEID', 'ADMITTED_DT', 'RACE', 'GENDER', 'AGE']
jail = jail[cols]

# -- rename columns for easy use
columns = ['id','date','race','gender','age']
jail.columns = columns

# -- drop rows with NaNs
jail = jail.dropna()

# -- change wrong datatypes
jail['age'] = jail['age'].astype(int)

# -- rename race and consolidate hispanic race
jail.loc[jail['race']=='B', 'race'] = 'black'
jail.loc[jail['race']=='O', 'race'] = 'hispanic'
jail.loc[jail['race']=='W', 'race'] = 'white'
jail.loc[jail['race']=='A', 'race'] = 'asian'
jail.loc[jail['race']=='I', 'race'] = 'native'
jail.loc[jail['race']=='U', 'race'] = 'unknown'

# -- change sex/gender from abbrev. to full name
jail.loc[jail['gender']=='M', 'gender'] = 'male'
jail.loc[jail['gender']=='F', 'gender'] = 'female'

jail
```

## **TIME SEIRES ANALYSIS**


### **Complaints Data**

```python
# -- group complaint data by date
complaint_group0 = complaint.groupby(complaint['date'].dt.to_period("M")).count()

# -- make plot showing the number of complaints in NYC (per month)
fig, ax = plt.subplots(figsize=(10,5))

complaint_group0.plot(y='id', ax=ax, legend=False, color='k', lw=1.5)
ax.set_xlabel('')
ax.set_ylabel('Total Number of Complaints in NYC', fontsize=10)
ax.grid(axis='x', which='major', lw=1)
ax.grid(axis='x', which='minor', lw=0.3)

fig.show()
```

A time series plot showing the total number of complaints in New York City (NYC) from 2012 to 2022. We can see some pattern in the plot/data, where the total number of complaints is high during the summer and lower during the winter. The lowest total number of complaints was recorded in March 2020. This is due the emergence of the COVID pandemic. From the plot, February seems to record the lowest total number of complaints almost every year. This is because February has the least number of days of all the months. To deal with this, there is a need to divide the total number by the number of days in each month. This will give a plot which is contains the average number of complaints per day.

```python
# -- divide grouped complaint data by number of days in each month
complaint_group = complaint_group0.divide(complaint_group0.index.days_in_month, axis=0)

# -- calculate 3-moth and 12-month rolling mean for the data
complaint_group_03 = complaint_group.rolling(3, center=True).mean()
complaint_group_12 = complaint_group.rolling(12, center=True).mean()

# -- make plot showing the average number of complaints in NYC per day in a month
fig, ax = plt.subplots(facecolor='whitesmoke', figsize=(10,5))

complaint_group.plot(y='id', ax=ax, color='k', lw=2, label='raw data')
complaint_group_03.plot(y='id', ax=ax, color='r', label='3-month rolling mean')
complaint_group_12.plot(y='id', ax=ax, color='b', label='12-month rolling mean')
ax.set_xlabel('')
ax.set_ylabel('Average Number of Complaints in NYC per day', fontsize=10)
ax.grid(axis='x', which='major', lw=1)
ax.grid(axis='x', which='minor', lw=0.3)

fig.show()
```

A time series plot showing the average number of complaints in NYC per day from 2012 to 2022. Compared to the prior plot, we now see that February is no longer recording the lowest values in all of the years. Low values were recorded in Nov 2012, Jan 2015 and March 2017. The red line shows the 3-month rolling mean of the complaints data time series. The blue line also shows the 12-month rolling mean of the time series plot. Generally, we can see a slight decrease in the average number complaints form 2012 to 2020. However, after the COVID 19 pandemic, that number has risen high, even higher that what it used to be in 2012. This confirms the findings made from the 2012, 2019 and 2022 cholopleth maps analysis.

Before delving into any calculations, a look at the histograms above shows a 'supposed' correspondence in the distribution in race in both the arrests and incarceration data. But the question is, 'IS THIS TRUE? ARE THE UNDERLYING DISPARITIES?

To answer this, we would have to first compare the rate of incarceration for each race before and after the COVID pandemic. That shows us if the rate of incarceration for eachrace has increased or decreased. To check for any racial disparities, the rates of incarceration for the race of people of colour will be compared to the rate of incarceration of the 'white' race.

```python
# -- subselect arrest data for cases in 2019 and remove rows where race is unknown
arrest_temp19 = arrest[(arrest['date']>'2018-12-31') & (arrest['date']<'2020-01-01')].copy()
arrest_temp19 = arrest_temp19[arrest_temp19['race']!='unknown']

# -- subselect incarceration data for cases in 2019 and remove rows where race is unknown
jail_temp19 = jail0[(jail0['date']>'2018-12-31') & (jail0['date']<'2020-01-01')].copy()
jail_temp19 = jail_temp19[jail_temp19['race']!='unknown']

# -- calculate rate of incarceration for whites
white19 = len(jail_temp19[jail_temp19['race']=='white'])/len(arrest_temp19[arrest_temp19['race']=='white'])
print('Rate of whites being incarcerated after arrest in 2019 = {:0.5f}'.format(white19))

# -- calculate rate of incarceration for blacks
black19 = len(jail_temp19[jail_temp19['race']=='black'])/len(arrest_temp19[arrest_temp19['race']=='black'])
print('Rate of blacks being incarcerated after arrest in 2019 = {:0.5f}'.format(black19))

# -- calculate rate of incarceration for hispanics
hispanic19 = len(jail_temp19[jail_temp19['race']=='hispanic'])/len(arrest_temp19[arrest_temp19['race']=='hispanic'])
print('Rate of hispanics being incarcerated after arrest in 2019 = {:0.5f}'.format(hispanic19))

# -- calculate rate of incarceration for asians
asian19 = len(jail_temp19[jail_temp19['race']=='asian'])/len(arrest_temp19[arrest_temp19['race']=='asian'])
print('Rate of asians being incarcerated after arrest in 2019 = {:0.5f}'.format(asian19))
```

The rate of incarceration for **whites** in 2019 was **4 out of 100,000 arrests**.

The rate of incarceration for **blacks** in 2019 was **37 out of 100,000 arrests**.

The rate of incarceration for **hispanics** in 2019 was **12 out of 100,000 arrests**.

The rate of incarceration for **asians** in 2019 was **17 out of 100,000 arrests**.

Clearly, we can already see a big gap that lies between the races of people of colour compared to whites. To get the magnitude of that difference, the odds ratio need to be calculated

```python
# -- calculate likelihood of being incarcerated if black compared to white in 2019
odds_ratio_black19 = black19/white19
print('Likelihood of being incarcerated if black, compared to white in 2019 = {:0.2f}'.format(odds_ratio_black19))

# -- calculate likelihood of being incarcerated if hispanic compared to white in 2019
odds_ratio_hispanic19 = hispanic19/white19
print('Likelihood of being incarcerated if hispanic, compared to white in 2019 = {:0.2f}'.format(odds_ratio_hispanic19))

# -- calculate likelihood of being incarcerated if asian compared to white in 2019
odds_ratio_asian19 = asian19/white19
print('Likelihood of being incarcerated if hispanic, compared to white in 2019 = {:0.2f}'.format(odds_ratio_asian19))
```

With the above, we can see that in 2019, it was 9.44 times more likely for a black person to be incarcerated compared to a white person. For hispanics and asians, they were 3.2 and 4.2 times likely to be incarcerated compared to whites respectively.

```python
# -- subselect arrest data for cases in 2022 and remove rows where race is unknown
arrest_temp22 = arrest[(arrest['date']>'2021-12-31') & (arrest['date']<'2023-01-01')].copy()
arrest_temp22 = arrest_temp22[arrest_temp22['race']!='unknown']

# -- subselect incarceration data for cases in 2022 and remove rows where race is unknown
jail_temp22 = jail0[(jail0['date']>'2021-12-31') & (jail0['date']<'2023-01-01')].copy()
jail_temp22 = jail_temp22[jail_temp22['race']!='unknown']

# -- calculate rate of incarceration for whites
white22 = len(jail_temp22[jail_temp22['race']=='white'])/len(arrest_temp22[arrest_temp22['race']=='white'])
print('Rate of whites being incarcerated after arrest in 2022 = {:0.4f}'.format(white22))

# -- calculate rate of incarceration for blacks
black22 = len(jail_temp22[jail_temp22['race']=='black'])/len(arrest_temp22[arrest_temp22['race']=='black'])
print('Rate of blacks being incarcerated after arrest in 2022 = {:0.4f}'.format(black22))

# -- calculate rate of incarceration for hispanics
hispanic22 = len(jail_temp22[jail_temp22['race']=='hispanic'])/len(arrest_temp22[arrest_temp22['race']=='hispanic'])
print('Rate of hispanics being incarcerated after arrest in 2022 = {:0.4f}'.format(hispanic22))

# -- calculate rate of incarceration for asians
asian22 = len(jail_temp22[jail_temp22['race']=='asian'])/len(arrest_temp22[arrest_temp22['race']=='asian'])
print('Rate of asians being incarcerated after arrest in 2022 = {:0.4f}'.format(asian22))
```

The rate of incarceration for **whites** in 2022 was **240 out of 100,000 arrests**.

The rate of incarceration for **blacks** in 2022 was **590 out of 100,000 arrests**.

The rate of incarceration for **hispanics** in 2022 was **390 out of 100,000 arrests**.

The rate of incarceration for **asians** in 2022 was **120 out of 100,000 arrests**.

```python
# -- calculate likelihood of being incarcerated if black compared to white in 2022
odds_ratio_black22 = black22/white22
print('Likelihood of being incarcerated if black, compared to white in 2022 = {:0.2f}'.format(odds_ratio_black22))

# -- calculate likelihood of being incarcerated if hispanic compared to white in 2022
odds_ratio_hispanic22 = hispanic22/white22
print('Likelihood of being incarcerated if hispanic, compared to white in 2022 = {:0.2f}'.format(odds_ratio_hispanic22))

# -- calculate likelihood of being incarcerated if asian compared to white in 2022
odds_ratio_asian22 = asian22/white22
print('Likelihood of being incarcerated if asian, compared to white in 2022 = {:0.2f}'.format(odds_ratio_asian22))
```
